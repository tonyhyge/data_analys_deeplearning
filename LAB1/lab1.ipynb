```python
import numpy as np
import pandas as pd
```

### Part 1:

```python
data = pd.read_csv('dulieuxettuyendaihoc.csv')

print("Thông tin dữ liệu:")
print(data.info())
print("\nMô tả dữ liệu:")
print(data.describe(include='all'))

qualitative_vars = ['GT', 'DT', 'KV', 'KT']

quantitative_vars = [col for col in data.columns if col not in qualitative_vars]

print("\nDanh sách biến định tính:", qualitative_vars)
print("Danh sách biến định lượng:", quantitative_vars)

for col in qualitative_vars:
    data[col] = data[col].astype('category')

for col in quantitative_vars:
    data[col] = pd.to_numeric(data[col], errors='coerce')

for col in quantitative_vars:
    mean_value = data[col].mean()
    data[col] = data[col].fillna(mean_value)


for col in qualitative_vars:
    mode_value = data[col].mode()[0]
    data[col] = data[col].fillna(mode_value)


print("\nSố lượng giá trị thiếu sau khi xử lý:")
print(data.isnull().sum())

bins = [0, 5, 7, 10]
labels = ['Yếu', 'Trung bình', 'Giỏi']

score_columns = ['T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
                 'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
                 'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6']

for col in score_columns:
    if col in data.columns:
        data[col + '_cat'] = pd.cut(data[col], bins=bins, labels=labels, include_lowest=True)

if set(['DH1', 'DH2', 'DH3']).issubset(data.columns):
    data['Tong_DH'] = data[['DH1', 'DH2', 'DH3']].sum(axis=1)
    print("\nTổng điểm thi đại học (Tong_DH) đã được tính.")

if 'Tong_DH' in data.columns:
    bins_result = [0, 15, 20, 30]
    labels_result = ['Không đỗ', 'Xét tuyển', 'Đỗ đại học']
    data['Ket_qua'] = pd.cut(data['Tong_DH'], bins=bins_result, labels=labels_result, include_lowest=True)
    print("\nBiến kết quả (Ket_qua) đã được tạo dựa trên tổng điểm thi.")

print("\nDữ liệu sau khi xử lý (5 dòng đầu tiên):")
print(data.head())

data.to_csv('dulieuxettuyendaihoc_processed.csv', index=False)
```

### Part 2:

```python
df = pd.read_csv('dulieuxettuyendaihoc.csv')

qualitative_vars = ['GT', 'DT', 'KV', 'KT']
quantitative_vars = [col for col in df.columns if col not in qualitative_vars]

for col in qualitative_vars:
    df[col] = df[col].astype('category')

print("Các biến định tính:", qualitative_vars)
print("Các biến định lượng:", quantitative_vars)
print("\n10 dòng đầu tiên của dữ liệu:")
print(df.head(10))
print("\n10 dòng cuối cùng của dữ liệu:")
print(df.tail(10))
print("\nThống kê dữ liệu thiếu cho cột 'DT':")
print(df['DT'].isnull().value_counts())
print("\nCác giá trị duy nhất trong cột 'DT':")
print(df['DT'].unique())

df['DT'] = df['DT'].cat.add_categories([0])
df['DT'] = df['DT'].fillna(0)

print("\nSau khi xử lý, thống kê dữ liệu thiếu cho cột 'DT':")
print(df['DT'].isnull().value_counts())
print("\nThống kê dữ liệu thiếu cho cột 'T1':")
print(df['T1'].isnull().value_counts())
print("\nCác giá trị duy nhất trong cột 'T1':")
print(df['T1'].unique())

mean_T1 = df['T1'].mean()
df['T1'] = df['T1'].fillna(mean_T1)

print("\nSau khi xử lý, thống kê dữ liệu thiếu cho cột 'T1':")
print(df['T1'].isnull().value_counts())

score_columns = ['T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
                 'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
                 'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6']

for col in score_columns:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())

if set(['T1','L1','H1','S1','V1','X1','D1','N1']).issubset(df.columns):
    df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
if set(['T2','L2','H2','S2','V2','X2','D2','N2']).issubset(df.columns):
    df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
if set(['T6','L6','H6','S6','V6','X6','D6','N6']).issubset(df.columns):
    df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10

def classify_grade(score):
    if score < 5.0:
        return 'Y'
    elif 5.0 <= score < 6.5:
        return 'TB'
    elif 6.5 <= score < 8.0:
        return 'K'
    elif 8.0 <= score < 9.0:
        return 'G'
    else:
        return 'XS'

if 'TBM1' in df.columns:
    df['XL1'] = df['TBM1'].apply(classify_grade)
if 'TBM2' in df.columns:
    df['XL2'] = df['TBM2'].apply(classify_grade)
if 'TBM3' in df.columns:
    df['XL3'] = df['TBM3'].apply(classify_grade)

def convert_to_US(score):
    return (score / 10) * 4

if 'TBM1' in df.columns:
    df['US_TBM1'] = df['TBM1'].apply(convert_to_US)
if 'TBM2' in df.columns:
    df['US_TBM2'] = df['TBM2'].apply(convert_to_US)
if 'TBM3' in df.columns:
    df['US_TBM3'] = df['TBM3'].apply(convert_to_US)

def evaluate_admission(row):
    if row['KT'] in ['A', 'A1']:
        avg_score = (row['DH1'] * 2 + row['DH2'] + row['DH3']) / 4
    elif row['KT'] == 'B':
        avg_score = (row['DH1'] + row['DH2'] * 2 + row['DH3']) / 4
    else:
        avg_score = (row['DH1'] + row['DH2'] + row['DH3']) / 3
    return 1 if avg_score >= 5.0 else 0

if set(['DH1', 'DH2', 'DH3']).issubset(df.columns) and 'KT' in df.columns:
    df['KQXT'] = df.apply(evaluate_admission, axis=1)

df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
print("\nĐã lưu dữ liệu đã xử lý vào file 'processed_dulieuxettuyendaihoc.csv'.")
```
